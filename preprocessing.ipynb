```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

WIDTH = 78
HEIGHT = 30

root = 'data'
img_folder = os.path.join(root, 'image')
processed_folder = os.path.join(root, 'processed')
index = 28
img_file = os.path.join(img_folder, 'captcha_' + str(index).zfill(4) + '.png')
```

```python
img = cv2.imread(img_file)
denoise_img = cv2.fastNlMeansDenoisingColored(img, None, 28, 28, 7, 21)
cropped_img = denoise_img[6:21, 8:, :]

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(denoise_img)
plt.subplot(133)
plt.imshow(cropped_img)
plt.show()
print(cropped_img.shape)
```

```python
gray_img = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh)
plt.show()
```

```python
processed_img_file = os.path.join(processed_folder, 'processed_' + str(index).zfill(4) + '.jpg')
cv2.imwrite(processed_img_file, thresh)
```
